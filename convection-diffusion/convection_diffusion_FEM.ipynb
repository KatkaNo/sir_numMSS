**Решение стационарного уравнения конвекции-диффузии на отрезке [0,1] с помощью метода конечных элементов (МКЭ).**
**
Задача Дирихл**е:

$Pe \frac{du}{dx} - \frac{d^2u}{dx^2} = 0 \quad \text{ в } [0,1]$

$u(0)=0,u(1)=1u(0) = 0, \quad u(1) = 1u(0)=0,u({e^{Pe} - 1}$



**Аналитическое решение**:

$u(x) = \frac{e^{Pe \, x} - 1}{e^{Pe} - 1}$

$\frac{du}{dx} = {Pe}\frac{e^{Pe \, x}}{e^{Pe} - 1}$

**Дискретная задача**

\begin{align*}
-\nu \frac{u_{i+1} - 2u_i + u_{i-1}}{h^2} + a\frac{u_{i+1} - u_{i-1}}{2h} &= f_i \\
\end{align*}

**Модифицированная дискретная задача:**

Метод SUPG (Stabilized Finite Element Method with Streamline Upwind Petrov-Galerkin) вводит дополнительные члены для стабилизации численного решения уравнений в частных производных.

\begin{align*}
-(\nu + \delta i a) \frac{u_{i+1} - 2u_i + u_{i-1}}{h^2} + a\frac{u_{i+1} - u_{i-1}}{2h} &= f_i - \delta_i a \frac{f_{i+1} - f_{i-1}}{2h} \\
\end{align*}

Запишем уравнение в виде:

\begin{align*}
-(\nu + \delta i a) \frac{1}{h^2}\begin{bmatrix} -2 & 1 & 0 & \ldots & \ldots \\ 1 & -2 & 1 & 0 & \ldots \\ 0 & 1 & -2 & 1 & 0 \\ \vdots & \vdots & \ddots & \ddots & \vdots \\ \vdots & \vdots & 0 & 1 & -2 \end{bmatrix} \begin{bmatrix} u_1 \\ u_2 \\ \vdots \\ \vdots \\ u_N \end{bmatrix} + a \frac{1}{2h} \begin{bmatrix} 0 & 1 & 0 & \ldots & \ldots \\ -1 & 0 & 1 & 0 & \ldots \\ 0 & -1 & 0 & 1 & 0 \\ \vdots & \vdots & \ddots & \ddots & \vdots \\ \vdots & \vdots & 0 & -1 & 0 \end{bmatrix} \begin{bmatrix} u_1 \\ u_2 \\ \vdots \\ \vdots \\ u_N \end{bmatrix} = \begin{bmatrix} f_1 - \delta_1 a \frac{f_2 - f_0}{2h} \\ f_2 - \delta_2 a \frac{f_3 - f_1}{2h} \\ \vdots \\ \vdots \\ f_N - \delta_N a \frac{f_{N+1} - f_{N-1}}{2h} \end{bmatrix}
 \end{align*}

**Вид полученного численного решения и сравнение с точным решением**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
```

*Ф-я для записи трёхдиагональной матрицы в 3 массива*

```python
def extract_diagonals(K_global):
    # Размерность матрицы
    n = K_global.shape[0]

    # Инициализация массивов для главной, верхней и нижней диагоналей
    main_diagonal = np.zeros(n)
    upper_diagonal = np.zeros(n)
    lower_diagonal = np.zeros(n)

    # Извлечение элементов диагоналей
    main_diagonal = np.diag(K_global)
    upper_diagonal = np.diag(K_global, k=1)
    lower_diagonal = np.diag(K_global, k=-1)

    upper_diagonal = np.append(upper_diagonal,0)
    lower_diagonal = np.insert(lower_diagonal, 0, 0)  

    return main_diagonal, upper_diagonal, lower_diagonal
```

*Метод прогонки*

```python
def run_through(aup, a, adown, d):
    L = len(a)
    alpha = [0] * L
    beta = [0] * L
    u = [0] * L

    alpha[0] = -(aup[0] / a[0])
    beta[0] = d[0] / a[0]

    # u[0] = alpha[0] * u[1] + beta[0]

    for l in range(1, L - 1):
        alpha[l] = -aup[l] / (a[l] + adown[l] * alpha[l - 1])
        beta[l] = (d[l] - adown[l] * beta[l - 1]) / (a[l] + adown[l] * alpha[l - 1])

    u[L - 1] = (d[L-1] - adown[L-1] * beta[L - 2]) / (a[L-1] + adown[L-1] * alpha[L - 2])

    for l in range(L-2, -1, -1):
        u[l] = alpha[l] * u[l+1] + beta[l]

    return u
```

*Аналитическое решение*

```python
def analytic(x, Pe):
    return (np.exp(Pe * x) - 1) / (np.exp(Pe) - 1)
```

*Инициализация системы: правая часть и матрица*

```python
def initialize_linear_system(L_m, delta, Pe, h):
    a, b, c, d = [0] * L_m, [0] * L_m, [0] * L_m, [0] * L_m

    c[0] = 0
    b[0] = -2 * (-1 - delta * Pe) / h**2
    a[0] = (-1 - delta * Pe) / h**2 + Pe / (2 * h)
    d[0] = 0

    c[L_m - 1] = (-1 - delta * Pe) / h**2 - Pe / (2 * h)
    b[L_m - 1] = -2 * (-1 - delta * Pe) / h**2
    a[L_m - 1] = 0
    d[L_m - 1] += -(-1 - delta * Pe) / h**2 - Pe / (2 * h)

    for i in range(1, L_m - 1):
        a[i] = (-1 - delta * Pe) / h**2 + Pe / (2 * h)
        b[i] = -2 * (-1 - delta * Pe) / h**2
        c[i] = (-1 - delta * Pe) / h**2 - Pe / (2 * h)
        d[i] = 0

    return a, b, c, d
```

*Основной цикл для визуализации*

```python
def solve_and_plot(Pe_values, Nodes):
    for Pe in Pe_values:
        for n in Nodes:
            x = np.linspace(0, 1, n + 1)
            analytical = analytic(x, Pe)
            h = 1 / n

            # delta
            delta = h / 2 * (1 / np.tanh(Pe * h / 2) - 2 / (Pe * h))

            L_m = n - 1
            a, b, c, d = initialize_linear_system(L_m, delta, Pe, h)

            u = run_through(a, b, c, d)
            u = np.concatenate([[analytical[0]], u, [analytical[-1]]])
            max_error_delta = max(abs(analytical - u))

            # delta = h
            delta = h

            L_m = n - 1
            a, b, c, d = initialize_linear_system(L_m, delta, Pe, h)

            u_h = run_through(a, b, c, d)
            u_h = np.concatenate([[analytical[0]], u_h, [analytical[-1]]])
            max_error_delta_h = max(abs(analytical - u_h))

            # delta = 0
            delta = 0

            L_m = n - 1
            a, b, c, d = initialize_linear_system(L_m, delta, Pe, h)

            u_0 = run_through(a, b, c, d)
            u_0 = np.concatenate([[analytical[0]], u_0, [analytical[-1]]])
            max_error_delta_0 = max(abs(analytical - u_0))

            max_numeric, min_numeric = max(u), min(u)
            max_numeric_h, min_numeric_h = max(u_h), min(u_h)
            max_numeric_0, min_numeric_0 = max(u_0), min(u_0)

            print(f'Pe={Pe}, n={n}, max_error={max_error_delta}')
            
            plt.figure(figsize=(4, 4)) 
            plt.plot(x, analytical, label='Analytic')
            plt.plot(x, u_h, label='Numerical (h)')
            plt.plot(x, u, label='Numerical (delta)')
            plt.plot(x, u_0, label='Numerical (delta=0)')
            plt.title(f'Graph of u(x): Pe = {Pe}, N = {n}')
            plt.xlabel('x')
            plt.ylabel('u')
            plt.legend()
            plt.grid(True)
            plt.show()
            plt.tight_layout()
```

```python
Pe_values = [0.001, 0.5, 1, 10, 50, 100, 150]
Nodes = [10, 50, 100, 500, 1000, 5000, 10000]
solve_and_plot(Pe_values, Nodes)
```

**Анализ скорости сходимости**

```python
from prettytable import PrettyTable

def solve_and_table(Pe_values, Nodes):
    table = PrettyTable()
    table.field_names = ["Pe", "n", "Максимальная ошибка (delta)", "Максимальная ошибка (delta=h)", "Максимальная ошибка (delta=0)"]

    for Pe in Pe_values:
        for n in Nodes:
            x = np.linspace(0, 1, n + 1)
            analytical = analytic(x, Pe)
            h = 1 / n

            # delta
            delta = h / 2 * (1 / np.tanh(Pe * h / 2) - 2 / (Pe * h))

            L_m = n - 1
            a, b, c, d = initialize_linear_system(L_m, delta, Pe, h)

            u = run_through(a, b, c, d)
            u = np.concatenate([[analytical[0]], u, [analytical[-1]]])
            max_error_delta = max(abs(analytical - u))

            # delta = h
            delta = h

            L_m = n - 1
            a, b, c, d = initialize_linear_system(L_m, delta, Pe, h)

            u_h = run_through(a, b, c, d)
            u_h = np.concatenate([[analytical[0]], u_h, [analytical[-1]]])
            max_error_delta_h = max(abs(analytical - u_h))

            # delta = 0
            delta = 0

            L_m = n - 1
            a, b, c, d = initialize_linear_system(L_m, delta, Pe, h)

            u_0 = run_through(a, b, c, d)
            u_0 = np.concatenate([[analytical[0]], u_0, [analytical[-1]]])
            max_error_delta_0 = max(abs(analytical - u_0))

            table.add_row([Pe, n, max_error_delta, max_error_delta_h, max_error_delta_0])

    print(table)
```

```python
solve_and_table(Pe_values, Nodes)
```

Метод SUPG позволяет избавиться от осцилляций независимо от Pe и шага сетки.
