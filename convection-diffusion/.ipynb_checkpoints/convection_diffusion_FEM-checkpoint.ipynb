```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import solve
```

```python
def extract_diagonals(K_global):
    # Размерность матрицы
    n = K_global.shape[0]

    # Инициализация массивов для главной, верхней и нижней диагоналей
    main_diagonal = np.zeros(n)
    upper_diagonal = np.zeros(n)
    lower_diagonal = np.zeros(n)

    # Извлечение элементов диагоналей
    main_diagonal = np.diag(K_global)
    upper_diagonal = np.diag(K_global, k=1)
    lower_diagonal = np.diag(K_global, k=-1)

    upper_diagonal = np.append(upper_diagonal,0)
    lower_diagonal = np.insert(lower_diagonal, 0, 0)  

    return main_diagonal, upper_diagonal, lower_diagonal
```

```python
def solver(aup, a, adown, d):
    L = len(a)
    alpha = [0] * L
    beta = [0] * L
    u = [0] * L

    alpha[0] = -(aup[0] / a[0])
    beta[0] = d[0] / a[0]

    # u[0] = alpha[0] * u[1] + beta[0]

    for l in range(1, L - 1):
        alpha[l] = -aup[l] / (a[l] + adown[l] * alpha[l - 1])
        beta[l] = (d[l] - adown[l] * beta[l - 1]) / (a[l] + adown[l] * alpha[l - 1])

    u[L - 1] = (d[L-1] - adown[L-1] * beta[L - 2]) / (a[L-1] + adown[L-1] * alpha[L - 2])

    for l in range(L-2, -1, -1):
        u[l] = alpha[l] * u[l+1] + beta[l]

    return u
```

```python
# Определение коэффициентов и граничных условий
a, b = 0, 1  # Изменено, чтобы соответствовать вашей области [0, 1]
Pe_values = [0, 0.5, 100]  # Числа Пекле

# Определение количества узлов и элементов
num_nodes = 100

def finite_element_solution(a, b, Pe, num_nodes):
    # Определение узлов и конечных элементов
    nodes = np.linspace(a, b, num_nodes)
    elements = np.array([range(i, i + 2) for i in range(num_nodes - 1)])

    # Построение матрицы жесткости для каждого элемента
    def local_stiffness_matrix(element, nodes):
        h = nodes[element[1]] - nodes[element[0]]
        k_matrix = np.array([[1, -1], [-1, 1]]) + Pe * np.array([[1, 0], [0, 1]]) * h / 2
        return k_matrix / h

    # Сборка глобальной матрицы жесткости
    K_global = np.zeros((num_nodes, num_nodes))

    for element in elements:
        K_local = local_stiffness_matrix(element, nodes)

        for i in range(2):
            for j in range(2):
                K_global[element[i], element[j]] += K_local[i, j]

    # Применение граничных условий
    K_global[0, 0] = 1
    K_global[0, 1:] = 0
    K_global[-1, -1] = 1
    K_global[-1, :-1] = 0

    # Решение системы линейных уравнений
    F_global = np.zeros(num_nodes)
    F_global[-1] = 1  # Граничное условие u(1) = 1
    #U = solve(K_global, F_global)
    a, aup, adown = extract_diagonals(K_global)
    U = solver(aup, a, adown, F_global)
    
    return U

# Построение графиков решений для разных чисел Пекле
for Pe in Pe_values:
    x_plot = np.linspace(a, b, num_nodes)
    U_p_plot = finite_element_solution(a, b, Pe, num_nodes)
    plt.plot(x_plot, U_p_plot, label=f'Pe={Pe}')

plt.xlabel('x')
plt.ylabel('u(x)')
plt.title('Число Пекле для разных Pe')
plt.legend()
plt.grid(True)
plt.show()
```
